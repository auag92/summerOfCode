```python
seed = 101
np.random.seed(seed)
Nx = 100
Ny = 100

X, Y = np.mgrid[0:Nx, 0:Ny]
dx = 0.1
dy = 0.07
lx = 3.4
ly = 4.2
theta_x = 2*np.pi*X*dx/lx
theta_y = 2*np.pi*Y*dy/ly
theta_o = 2*np.pi*(X+3*Y)*dy/ly
f = np.sin(theta_x) + np.sin(theta_y) + np.sin(theta_o)
draw(X)
draw(Y)
draw(f)
```

```python
import numpy.fft as fft
s = (Nx,Ny)
f1_ = fft.fftn(f,s = s)
f1_ = fft.fftshift(f1_)
print(f1_.shape)
draw(abs(f1_)/np.sqrt(f.size))
f_ = fft.ifftn(f1_)
# f_ = fft.ifftshift(f_)
draw(abs(f_))
```
